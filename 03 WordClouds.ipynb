```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import matplotlib as mpl

from cleaning import cf
```

```python
data = pd.read_csv('emotion_tweets_cleaned.csv')
data['emoji'] = data.tweet.apply(lambda x: cf.extract_emojis(x))

data['tweet'] = data.tweet.apply(lambda x: cf.remove_specific_links(x))
data['tweet'] = data.tweet.apply(lambda x: cf.remove_urls(x))
data['tweet'] = data.tweet.apply(lambda x: cf.remove_punctuation(x))


data['tweet'] = data.apply(lambda x: cf.add_emojis(x), axis=1)

data
```

```python
cmap = plt.cm.Blues(np.linspace(0,1,20))
cmap = mpl.colors.ListedColormap(cmap[2:,:-1])
```

```python
feeling = 'fear'
plt.figure(figsize=(10,7))
text = ' '.join(data[data.emotion == feeling].tweet)
plt.imshow(WordCloud(max_font_size=80,background_color='white',colormap = cmap,font_path='arial').generate(text), interpolation="bilinear")
plt.gca().set_facecolor('white')
plt.axis("off")

plt.savefig(feeling+'.png', dpi=400)
```
